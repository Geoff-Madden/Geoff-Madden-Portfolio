There is no missing data so we will use all the rows

Calculate the fuel cost of driving 10000 miles each year for three years.

Create a column to calculate total litres based on the mpg.

```python
df_combined['total litres']=30000/df_combined['mpg']*4.546
df_combined['total litres'] = df_combined['total litres'].astype(int)
df_combined.head()
```

How much will my car depreciate?
If you estimated the typical lifespan of a car to be 20 years, the average depreciation every year would actually be a 15% drop every 12 months. https://motorway.co.uk/sell-my-car/guides/car-depreciation-guide

Add a column to estimate the depreciation of the vehicle over three years

Using the formula for compound interest we can see that over three years the depreciation cost will amount to 38.58% of the purchase price.

```python
df_combined['3yr depreciation cost']=df_combined['price'] * .3858
df_combined['3yr depreciation cost'] = df_combined['3yr depreciation cost'].round(2)
df_combined.head()
```

Create a column called 3yr fuel costs based on fuel type and total litres

Examine the different fuel types

```python
fuel_types = df_combined['fueltype'].unique()
fuel_types
```

We do not have enough data to consider options other than petrol or diesel

```python
df_combined = df_combined[df_combined['fueltype'].isin(['Petrol', 'Diesel'])]
df_combined.head()
```

```python
prices = {'Petrol': 145.3, 'Diesel': 154.6}
df_combined['3yr fuel cost'] = df_combined.apply(lambda row: row['total litres'] * prices.get(row['fueltype'], 1), axis=1)/100
df_combined['3yr fuel cost'] = df_combined['3yr fuel cost'].round(2)
df_combined.head()    
```

We will base the total cost of ownership over three years on tax, 3yr depreciation costs and 3yr fuel costs

```python
df_combined['total cost'] = df_combined[['tax', '3yr depreciation cost', '3yr fuel cost']].sum(axis=1)
df_combined.head()  
```

Finally we will assume that the cheapest insurance groups will relate to the smaller engine sizes

```python
df_combined['enginesize'].describe()
```

Remove all engine size values of 0

```python
df_filtered = df_combined[df_combined['enginesize'] != 0]
df_filtered.head()
```

```python
df_filtered['enginesize'].describe()
```

```python
df_filtered = df_filtered[df_filtered['enginesize'] == 1]
df_filtered.head()
```

Find the top 20 cars with the minimum total cost

```python
df_lowest_cost = df_filtered.nsmallest(20, columns = 'total cost')
df_lowest_cost
```

We can group the cars by make, model year and price for the labels

```python
df_lowest_cost['grouped'] = df_lowest_cost.apply(lambda row: f"{row['make']}_{row['model']}_{row['year']}_{row['price']}", axis=1)
df_lowest_cost
```

We can display this information on a horizontal bar chart.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='total cost', y='grouped', data=df_lowest_cost, orient='h', color = 'steelblue')
plt.xlabel('Total Cost')
plt.ylabel('Car by make, model, year and price')
plt.title('Cheapest cars by overall cost over three years')
plt.show()
```
